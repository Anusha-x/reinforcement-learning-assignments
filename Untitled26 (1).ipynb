```python
import numpy as np

# States
states = ["NoPerson", "Far", "Near"]
n_states = len(states)

# Actions
actions = ["Open", "Close", "Wait"]
n_actions = len(actions)

# Rewards matrix [state][action]
rewards = np.array([
    [-1, 2, 0],   # NoPerson
    [0, 1, 2],    # Far
    [5, -2, 1]    # Near
])

# Transition probabilities P[s][a][s']
P = np.zeros((n_states, n_actions, n_states))

# Define transitions
# NoPerson
P[0][0] = [1, 0, 0]  # Open
P[0][1] = [1, 0, 0]  # Close
P[0][2] = [0.7, 0.3, 0]  # Wait

# Far
P[1][0] = [0, 0.6, 0.4]
P[1][1] = [0, 0.8, 0.2]
P[1][2] = [0, 0.5, 0.5]

# Near
P[2][0] = [0, 0, 1]
P[2][1] = [0, 0.3, 0.7]
P[2][2] = [0, 0.4, 0.6]

gamma = 0.9

# Initialize policy randomly
policy = np.random.choice(n_actions, size=n_states)

def policy_evaluation(policy):
    V = np.zeros(n_states)
    while True:
        delta = 0
        for s in range(n_states):
            v = V[s]
            a = policy[s]
            V[s] = sum(P[s][a][s_next] * (rewards[s][a] + gamma * V[s_next])
                       for s_next in range(n_states))
            delta = max(delta, abs(v - V[s]))
        if delta < 1e-4:
            break
    return V

def policy_improvement(V):
    new_policy = np.zeros(n_states, dtype=int)
    for s in range(n_states):
        Q = []
        for a in range(n_actions):
            q = sum(P[s][a][s_next] * (rewards[s][a] + gamma * V[s_next])
                    for s_next in range(n_states))
            Q.append(q)
        new_policy[s] = np.argmax(Q)
    return new_policy

# Policy Iteration
while True:
    V = policy_evaluation(policy)
    new_policy = policy_improvement(V)
    if np.array_equal(policy, new_policy):
        break
    policy = new_policy

print("Optimal Policy:")
for i, s in enumerate(states):
    print(f"{s} -> {actions[policy[i]]}")
```
