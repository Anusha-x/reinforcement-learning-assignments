```python
import numpy as np
import random

# Grid size
grid_size = 4

# Actions: up, down, left, right
actions = ["UP", "DOWN", "LEFT", "RIGHT"]
n_actions = 4

# Q-table
Q = np.zeros((grid_size, grid_size, n_actions))

# Rewards
goal = (3, 3)
penalty = [(1,1), (2,2)]

def get_reward(state):
    if state == goal:
        return 10
    elif state in penalty:
        return -10
    else:
        return -1

def step(state, action):
    x, y = state

    if action == 0: x -= 1
    elif action == 1: x += 1
    elif action == 2: y -= 1
    elif action == 3: y += 1

    x = max(0, min(grid_size-1, x))
    y = max(0, min(grid_size-1, y))

    return (x, y)

# Parameters
alpha = 0.1
gamma = 0.9
epsilon = 0.2
episodes = 500

# Training
for ep in range(episodes):
    state = (0,0)

    while state != goal:
        if random.uniform(0,1) < epsilon:
            action = random.randint(0,3)
        else:
            action = np.argmax(Q[state[0], state[1]])

        next_state = step(state, action)
        reward = get_reward(next_state)

        old_value = Q[state[0], state[1], action]
        next_max = np.max(Q[next_state[0], next_state[1]])

        Q[state[0], state[1], action] = old_value + alpha * (
            reward + gamma * next_max - old_value
        )

        state = next_state

# Testing
state = (0,0)
path = [state]

while state != goal:
    action = np.argmax(Q[state[0], state[1]])
    state = step(state, action)
    path.append(state)

print("Learned Path:")
print(path)
```
